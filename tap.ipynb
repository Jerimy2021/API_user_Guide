```python
import requests
import urllib.parse
import csv
from math import cos, sin,radians

class FetchPlanetsError(Exception):
    pass

def fetch_name_planets(system_solar):
    encoded_system_solar = urllib.parse.quote_plus(system_solar)
    api_url = f"https://exoplanetarchive.ipac.caltech.edu/TAP/sync?query=select+distinct(pl_name),sy_dist,ra,dec+from+ps+where+pl_name+like+%27%25{encoded_system_solar}%25%27+order+by+pl_name+desc&format=csv"
    headers = {
        'User-Agent': 'App'
    }
    response = requests.get(api_url, headers=headers)


    response_data = list(csv.reader(response.text.splitlines()))
    
    if not response_data or len(response_data) == 1:  
        raise FetchPlanetsError("Error 404: No data found for the specified system solar")
    
    return response_data

# Ejemplo de uso
sistema_solar = '14 Her'
try:
    planets_csv = fetch_name_planets(sistema_solar)[1:]
    print(planets_csv)

except FetchPlanetsError as e:
    print(e)
```

```python
import requests
import urllib.parse
import csv


def fetch_stellar_hosts():
    api_url = 'https://exoplanetarchive.ipac.caltech.edu/TAP/sync?query=select+distinct(sy_name)+from+stellarhosts&format=csv'
    headers = {
        'User-Agent': 'App'
    }
    response = requests.get(api_url, headers=headers)
    if response.status_code == 404:
        raise FetchPlanetsError("Error 404: Resource not found")
    
    response_data = list(csv.reader(response.text.splitlines()))
    
    return response_data[1:]


print(fetch_stellar_hosts())



```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from io import BytesIO
from PIL import Image
import numpy as np

# URL de la pgina web que contiene las texturas
url = "https://www.solarsystemscope.com/textures/"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Encuentra todos los enlaces de descarga de texturas
    textures = soup.find_all('a', class_='btn-type-1-blue texture-download-button')
    
    # Imprime la informacin de las texturas
    for texture in textures:
        download_link = texture['href']
        if '2k' in download_link:
            # Construye la URL completa concatenando la parte final de la URL con la URL base de la pgina
            full_download_link = urljoin(url, download_link)
            # Descarga la imagen desde el enlace completo
            image_response = requests.get(full_download_link)
            if image_response.status_code == 200:
                # Cargar la imagen desde el contenido de la respuesta
                image = Image.open(BytesIO(image_response.content))
                # Convertir la imagen a una matriz numpy
                image_array = np.array(image)
                # Ahora puedes trabajar con image_array
                print(f"Dimensiones de la imagen: {image_array.shape}")
                # Por ejemplo, imprimir los primeros 5 pxeles de la imagen
                print("Primeros 5 pxeles de la imagen:")
                print(image_array[:5])
            else:
                print(f"Error al descargar imagen desde {full_download_link}, estado: {image_response.status_code}")

else:
    print(f"Failed to retrieve the page, status code: {response.status_code}")
```

```python
import re
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from io import BytesIO
from PIL import Image
import numpy as np
def extract_planet_name(filename):
    match = re.search(r'2k_(.+)\.jpg', filename)
    if match:
        return match.group(1)
    return None

def fetch_textures_planets(planet_id):
    url = "https://www.solarsystemscope.com/textures/"
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        textures = soup.find_all('a', class_='btn-type-1-blue texture-download-button')

        for texture in textures:
            download_link = texture['href']
            if '2k' in download_link:
                name_planet = extract_planet_name(download_link)
                if name_planet == planet_id:
                    print(name_planet)
                    full_download_link = urljoin(url, download_link)
                    print(full_download_link)
                    image_response = requests.get(full_download_link)
                    if image_response.status_code == 200:
                        return image_response
                    
print(fetch_textures_planets('mercury'))
```
